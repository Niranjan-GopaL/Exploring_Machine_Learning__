```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('q1_1.csv')
df
```

```python
# Plot V_out vs V_in

plt.plot(df['V(in)'],df['V(vout)'],'-o')
plt.xlabel("Vin(Volt)")
plt.ylabel("Vout(Volt)")
plt.title('Vout vs Vin for 45nm',weight = 'bold')
plt.grid('on')
```

```python
# Plot Id vs V_in (NMOS)
plt.plot(df['V(in)'],df['Id(M1)'],'-o')
plt.xlabel("Vin(Volt)")
plt.ylabel("Id(A)")
plt.title('Id vs Vin 45nm (NMOS)',weight = 'bold')
plt.grid('on')
```

```python
# Plot Is vs V_in (PMOS)
plt.plot(df['V(in)'],df['Is(M2)'],'-o')
plt.xlabel("Vin(Volt)")
plt.ylabel("Id(A)")
plt.title("Is vs Vin 45nm(PMOS)",weight = 'bold')
plt.grid('on')
```
